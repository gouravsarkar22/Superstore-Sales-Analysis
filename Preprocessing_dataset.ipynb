### Superstore Sales Analysis

```python
# import necessary libraries
import pandas as pd
```

```python
# load data
df = pd.read_csv('Superstore Dataset.csv')
df.head()
```

```python
df.info()
```

```python
# Convert 'Order Date' and 'Ship Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
# Sort by 'Order Date'
df = df.sort_values(by='Order Date')
df.head()
```

```python
# drop rows where ship date is before order date
df = df[df['Ship Date'] >= df['Order Date']]
df.head()
```

```python
# Find the number of missing values in each column
df.isnull().sum()
```

```python
# Find the number of duplicate rows in the dataset
df.duplicated().sum()
```

```python
# Show duplicate rows
df[df.duplicated(keep = 'last')]
```

```python
# double check the duplicated rows
df[df.duplicated(keep = 'first')]
```

```python
#Remove duplicates
df=df.drop_duplicates()
df.info()
```

```python
# Save the preprocessed dataset
df.to_csv('Superstore_Dataset_Preprocessed.csv', index=False)
```
